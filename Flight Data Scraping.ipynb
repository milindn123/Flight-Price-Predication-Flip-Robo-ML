```python
from selenium.webdriver.common.keys import Keys

#we are scrollling and loading the whole page then we will pass it to Beautiful soup

driver.get("https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=DEL&destinationCountry=IN&flexi=0&flight_depart_date=19%2F11%2F2021&hb=&noOfSegments=1&origin=MAA&originCountry=IN&type=O&unique=1634561596448&viewName=normal")

time.sleep(4)

element = driver.find_element_by_tag_name('body')

while True:
       
    try:
        element.send_keys(Keys.PAGE_DOWN)
        time.sleep(3)
    except: 
        break

```

```python
soup=BeautifulSoup(driver.page_source, 'lxml')

arrtimes = soup.find_all('p', attrs={'autom': 'arrivalTimeLabel'})
deptimes = soup.find_all('div', attrs={'autom': 'departureTimeLabel'})
duration = soup.find_all('p', attrs={'autom': 'durationLabel'})
price = soup.find_all('div', attrs={'class': 'i-b tipsy fare-summary-tooltip fs-18'})
flight = soup.find_all('span', attrs={'class': 'i-b text ellipsis'})
stops = soup.find_all('div', attrs={'class': 'font-lightgrey fs-10 tipsy i-b fs-10'}) 


Stops = []
for st in stops:
    for n in st.select('span'):
        Stops.append(n.getText())  
        
    
Fligh_Name = []
for f in flight:
    Fligh_Name.append(f.getText())
    
deptime = []
for div in deptimes:
    deptime.append(div.getText())
    
    
arrtime = []
for div in arrtimes:
    arrtime.append(div.getText())   

durtimes = []
for p in duration:
    durtimes.append(p.getText()) 
    
Price = []
for pr in price:
    Price.append(pr.getText()) 
        
```

```python

source = soup.find_all('p', attrs={'class': 'fs-10 font-lightgrey no-wrap city ellipsis'}) 

Source = []
for sr in source:
    labels = (str(sr.text)) 
    Source.append(labels)
        
        
```

```python

Via = soup.find_all('td', attrs={'class': 'text-left pr-5'}) 

Route = []
for ro in Via:
    if ro.getText() =='Stop':
        pass
    else:
        Route.append(ro.getText() if Via else "")
    
#result = soup.find('div', class_="something")
#data['val1'].append(result.text if result else "N/A")
```

```python
for i in range(45):
    heading =driver.find_element_by_xpath('//*[@id="jssor_1"]/div/div/div/div/div[2]/div[7]/div[1]/div[1]/div[1]')
    heading.click()
    time.sleep(6)
    
    

    #Airline
    for i in driver.find_elements_by_xpath("//*[@id='ResultDiv']/div/div/div/div/div/div/div/div/div/span[1]"):
        Airline.append(i.text)


    #departure time
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[2]/span[1]'):
        Dep_time.append(i.text)



    #arrival time
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[4]/span[1]'):
        Arr_time.append(i.text)


    #duration in hours
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[3]/span[1]'):
        Durtion.append(i.text)


    #route 
    count=0

    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[3]/span[2]'):
        Stops.append(i.text)
        detail =driver.find_element_by_xpath(f"//*[@id='{count}']")
        detail.click()
        route1 =driver.find_element_by_xpath(f'//*[@id="fd{count}"]/div/div[2]/div[1]/span[1]/span[1]')
        route2 =driver.find_element_by_xpath(f'//*[@id="fd{count}"]/div/div[2]/div[1]/span[1]/span[2]')
        """try:
            route3 =driver.find_element_by_xpath(f'//*[@id="fd{count}"]/div/div[2]/div[2]/span[1]/span[2]')
        except:
            break"""



        str1 = '--->'

        via =route1.text + str1
        route4= via + route2.text
        # route5 =route4 + str1 
        # route6 =route5 + route3.text
        Route.append(route4)


        count +=1

    #source
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div[1]/div[2]/span[2]'):
        Source.append(i.text)
    #destination
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[4]/span[2]'):
        Destination.append(i.text)
    #price    
    for i in driver.find_elements_by_xpath('//*[@class="col-md-8 col-sm-8 col-xs-9 txt-r6-n ng-binding"]'):
        Price.append(i.text)
    #date
    i  = driver.find_elements_by_xpath('//*[@class="col-md-8 col-sm-8 col-xs-9 txt-r6-n ng-binding"]')
    for j in i:
        date1 =driver.find_element_by_xpath('//*[@id="jssor_1"]/div/div/div/div/div[2]/div[6]/div[1]/div[1]/div[1]') 
        Date.append(date1.text)   


```
